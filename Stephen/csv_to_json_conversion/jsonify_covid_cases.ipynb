```python
import pandas as pd
import json
from pathlib import Path
```

```python
# Load the CSV file into a DataFrame
file_path = Path('../johns_hopkins_data_merge/main_covid_cases_2020-2023_merge.csv')
covid_df = pd.read_csv(file_path)

# Print out the column names to verify
print("Columns in the DataFrame:", covid_df.columns.tolist())
```

```python
covid_cases = []
# Iterate over the DataFrame rows
for index, row in covid_df.iterrows():
    # Create a dictionary with the relevant information
    covid_dict = row.to_dict()
    # Append the dictionary to the list
    covid_cases.append(covid_dict)

# Convert the list of dictionaries to a JSON-formatted string
covid_list_json = json.dumps(covid_cases, indent=4)

# Print the JSON-formatted string
print(covid_list_json)
```

```python
# Save the JSON data to a file
json_output_path = Path('./covid_cases.json')
with open(json_output_path, 'w') as json_file:
    json_file.write(covid_list_json)

print(f"JSON data has been written to {json_output_path}")
```
