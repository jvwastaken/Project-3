```python
#Dependency
import pandas as pd

#Included the two columns of interest: States and 2023 Total Estimated Sales
us_gun_sales = {
    'Province_State': ['Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California', 'Colorado', 'Connecticut', 'Delaware', 'District of Columbia', 'Florida', 'Georgia', 
                       'Hawaii', 'Idaho', 'Illinois', 'Indiana', 'Iowa', 'Kansas', 'Kentucky', 'Louisiana', 'Maine','Maryland', 'Massachusetts', 'Michigan', 'Minnesota', 
                       'Mississippi', 'Missouri', 'Montana', 'Nebraska', 'Nevada', 'New Hampshire', 'New Jersey', 'New Mexico', 'New York', 'North Carolina', 'North Dakota', 
                       'Ohio', 'Oklahoma', 'Oregon', 'Pennsylvania', 'Rhode Island', 'South Carolina', 'South Dakota', 'Tennessee', 'Texas', 'Utah', 'Vermont', 'Virginia', 'Washington', 
                       'West Virginia', 'Wisconsin', 'Wyoming'],

    '2023 Total Estimated Sales': ["487,077", "80,341", "407,288", "200,279", "1,043,421", "443,800", "120,044", "47,080", "4,351", "1,316,471", "418,452", 
                                   "23,528", "160,120", "504,452", "498,169", "82,795", "179,350", "288,877", "282,155", "118,297", "213,437", "136,398", "555,650", "309,425", 
                                   "227,199", "520,488", "134,008", "26,987", "118,998", "136,502", "175,711", "163,624", "337,814", "499,561", "66,933", 
                                   "583,314", "325,700", "446,556", "841,523", "28,555", "259,145", "72,303", "633,015", "1,347,589", "143,430", "46,518", "581,698", "489,551", 
                                   "163,057", "413,680", "65,102"]
}

#Convert to DataFrame
us_gun_sales_df = pd.DataFrame(us_gun_sales)

#Cleanse and convert the '2023 Total Estimated Sales' column to numeric values
us_gun_sales_df['2023 Total Estimated Sales'] = us_gun_sales_df['2023 Total Estimated Sales'].str.replace(',', '').str.replace('"', '').astype(int)

us_gun_sales_df
```

```python
#Export CSV file
us_gun_sales_df.to_csv('US_gun_sales_data.csv', index=False)
```

```python
#Adding in a new column: Guns per Capita
us_population_df = pd.read_csv("US_population.csv", thousands=',')
# Drop null columns 3 to 7 
us_population_df = us_population_df.drop(us_population_df.columns[2:8], axis=1)
us_population_df
```

```python
#use merge to combine dataframes by Province_State
us_population_and_gun_sales_df = pd.merge(us_gun_sales_df, us_population_df, on="Province_State")
us_population_and_gun_sales_df
```

```python
# Dividing between guns sales over population and create new column: Guns per capita
us_population_and_gun_sales_df['Guns Per Capita'] = us_population_and_gun_sales_df['2023 Total Estimated Sales'] / us_population_and_gun_sales_df['Population']

# Formatting the 'Guns Per Capita' column as a percentage
us_population_and_gun_sales_df['Guns Per Capita'] = us_population_and_gun_sales_df['Guns Per Capita'].map('{:.2%}'.format)

# Removes % sign while converting the values to float for smooth merges in the final dataframe
us_population_and_gun_sales_df['Guns Per Capita'] = us_population_and_gun_sales_df['Guns Per Capita'].str.rstrip('%').astype('float').round(2)

# Remove "Population" column as it's unneeded
us_gun_sales_and_capita_df = us_population_and_gun_sales_df.drop(['Population'], axis='columns')

us_gun_sales_and_capita_df
```
